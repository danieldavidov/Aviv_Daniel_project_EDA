```python
import pandas as pd
import zipfile

zip_path = r'C:\Users\User\פרויקט ניתוח נתונים\archive.zip'

with zipfile.ZipFile(zip_path, 'r') as z:
    print(z.namelist())  # כדי לראות איזה קבצים יש בתוך הקובץ
    with z.open(z.namelist()[0]) as f:  # מניחים שהקובץ הראשון הוא ה-CSV
        df = pd.read_csv(f, encoding='cp1255')

df.head()
```
